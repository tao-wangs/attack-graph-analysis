```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import torch 

df = pd.read_csv('epss_scores.csv', skiprows=1)

df
```

```python
# Find the distribution of EPSS probabilities

bins = np.arange(0., 1.1, 0.1)
labels = ['0-0.1', 
          '0.1-0.2', 
          '0.2-0.3', 
          '0.3-0.4', 
          '0.4-0.5', 
          '0.5-0.6', 
          '0.6-0.7', 
          '0.7-0.8', 
          '0.8-0.9', 
          '0.9-1']

df['binned'] = pd.cut(df['epss'], bins=bins, labels=labels)
df['binned'].value_counts(sort=False)
```

```python
# counts = df['binned'].value_counts(sort=False)
# sns.barplot(x=counts.index, y=counts.values)

import matplotlib.pyplot as plt 

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

counts = df['binned'].value_counts(sort=False)

sns.barplot(x=counts.index, y=counts.values, alpha=0.5, label='EPSS Data')

plt.title('Distribution of EPSS Data')
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the EPSS scores dataset
file_path = 'epss_scores.csv'  # Update this with your file path
epss_data = pd.read_csv(file_path)

# Extracting the EPSS scores column
epss_scores = epss_data.iloc[:, 1].values

# Setting a style
sns.set(style="whitegrid")

# Plotting the histogram
plt.figure(figsize=(12, 8))
sns.histplot(epss_scores, bins=30, kde=True, color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Distribution of EPSS Scores', fontsize=20, fontweight='bold')
plt.xlabel('EPSS Score', fontsize=15)
plt.ylabel('Frequency', fontsize=15)

# Customizing ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()
```
