```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
full_dataset = df.epss.values 
size = len(full_dataset)

reduced_dataset = df[df['epss'] > 0.05].epss.values
size = len(reduced_dataset)

sns.set_theme()

# Full dataset plot
fig, axs = plt.subplots(2, 1, figsize=(6, 6), gridspec_kw={'height_ratios': [1, 4]})

sns.boxplot(x=full_dataset, ax=axs[0], width=0.3)
sns.histplot(full_dataset, bins=50, ax=axs[1], kde=False)

axs[1].set_title('')
axs[1].set_xlabel('EPSS Score')
axs[1].set_ylabel('Frequency Density')

plt.tight_layout()
plt.savefig('full_dataset_plot.png')
plt.show()

# Reduced dataset plot
fig, axs = plt.subplots(2, 1, figsize=(6, 6), gridspec_kw={'height_ratios': [1, 4]})

sns.boxplot(x=reduced_dataset, ax=axs[0], width=0.3)
sns.histplot(reduced_dataset, bins=50, ax=axs[1], kde=False)

axs[1].set_title('')
axs[1].set_xlabel('EPSS Score')
axs[1].set_ylabel('Frequency Density')

plt.tight_layout()
plt.savefig('reduced_dataset_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generating sample data with a known distribution
np.random.seed(42)
full_dataset = np.concatenate([np.random.normal(0, 1, 100000), np.random.normal(5, 1, 10000)])

fig, axs = plt.subplots(2, 1, figsize=(8, 10), gridspec_kw={'height_ratios': [1, 3]})

# Narrower box plot
sns.boxplot(x=full_dataset, ax=axs[0], width=0.3)

# Matching histogram
sns.histplot(full_dataset, bins=100, ax=axs[1], kde=False)

axs[0].set_title('Box Plot of Full Dataset')
axs[1].set_title('Histogram of Full Dataset')
axs[1].set_xlabel('Data Value')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
