```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the file
with open('50_comp_times.txt', 'r') as f:
    times = [float(line.strip()) for line in f]

# Create an array for the indices
nodes = np.arange(50, 2501, 50)

# Create the plot
# plt.figure(figsize=(10, 10))
plt.plot(nodes, times, marker='x')

# Add title and labels
plt.title('Compositional Analysis')
plt.xlabel('Number of nodes')
plt.ylabel('Inference time (s)')
plt.tight_layout()

# Show the plot
plt.show()
```
