```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the file
with open('max_parents_2/25_comp_times.txt', 'r') as f:
    times1 = [float(line.strip()) for line in f]

with open('max_parents_2/50_comp_times.txt', 'r') as f:
    times2 = [float(line.strip()) for line in f]

with open('max_parents_2/100_comp_times.txt', 'r') as f:
    times3 = [float(line.strip()) for line in f]

with open('max_parents_2/200_comp_times.txt', 'r') as f:
    times4 = [float(line.strip()) for line in f]

# Create an array for the indices
nodes1 = np.arange(25, 2501, 25)
nodes2 = np.arange(50, 2501, 50)
nodes3 = np.arange(100, 2501, 100)
nodes4 = np.arange(200, 2501, 200)

# Create the plot
# plt.figure(figsize=(10, 10))
plt.plot(nodes1, times1, marker='x', label='25 nodes')
plt.plot(nodes2, times2, marker='x', label='50 nodes')
plt.plot(nodes3, times3, marker='x', label='100 nodes')
plt.plot(nodes4, times4, marker='x', label='200 nodes')

# Add title and labels
plt.title('Compositional Analysis')
plt.xlabel('Number of nodes')
plt.ylabel('Inference time (s)')
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
